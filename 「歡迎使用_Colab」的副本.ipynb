舉例來說，以下是包含簡短 Python 指令碼的<strong>程式碼儲存格</strong>，可進行運算、將值儲存至變數中並列印運算結果：

```
# 安裝套件
!pip install -q transformers
!pip install -q ftfy regex tqdm
!pip install -q git+https://github.com/openai/CLIP.git
!pip install -q gtts
!pip install -q Pillow

# 匯入模組
import torch
import clip
from PIL import Image
from gtts import gTTS
from IPython.display import Audio, display
import uuid
from google.colab import files

# 載入 CLIP 模型
device = "cuda" if torch.cuda.is_available() else "cpu"
model, preprocess = clip.load("ViT-B/32", device=device)

# 建立畫作資料庫
known_paintings = {
    "Mona Lisa": {
        "artist": "Leonardo da Vinci",
        "year": "1503–1506",
        "description": "《蒙娜麗莎》是達文西於文藝復興時期創作的著名肖像畫，現藏於法國羅浮宮博物館。"
    },
    "Starry Night": {
        "artist": "Vincent van Gogh",
        "year": "1889",
        "description": "《星夜》是梵谷創作的名畫之一，展現他對夜空的幻想與內心的情感。"
    },
    "The Scream": {
        "artist": "Edvard Munch",
        "year": "1893",
        "description": "《吶喊》是孟克於1893年創作的表現主義畫作，描繪一種極端焦慮的情緒。"
    },
    "Girl with a Pearl Earring": {
        "artist": "Johannes Vermeer",
        "year": "1665",
        "description": "《戴珍珠耳環的少女》被譽為“北方的蒙娜麗莎”，是維梅爾最著名的作品之一。"
    },
    "The Persistence of Memory": {
        "artist": "Salvador Dalí",
        "year": "1931",
        "description": "《記憶的堅持》是達利的超現實主義名作，描繪了融化的時鐘，象徵時間的相對性。"
    },
    "The Night Watch": {
        "artist": "Rembrandt",
        "year": "1642",
        "description": "《夜巡》是倫勃朗的代表作之一，描繪了阿姆斯特丹民兵隊的集會場景。"
    },
    "The Birth of Venus": {
        "artist": "Sandro Botticelli",
        "year": "1486",
        "description": "《維納斯的誕生》是博提切利的名作，描繪了愛與美的女神維納斯誕生的場景。"
    },
    "The Creation of Adam": {
        "artist": "Michelangelo",
        "year": "1512",
        "description": "《亞當的創造》是米開朗基羅在西斯廷禮拜堂的天花板畫作，描繪了上帝賦予亞當生命的瞬間。"
    },
    "Guernica": {
        "artist": "Pablo Picasso",
        "year": "1937",
        "description": "《格爾尼卡》是畢卡索反戰代表作，表現西班牙內戰的暴力與痛苦。"
    },
    "The Kiss": {
        "artist": "Gustav Klimt",
        "year": "1907–1908",
        "description": "《親吻》是克里姆特的黃金時期作品，表現愛與親密的主題。"
    },
    "The School of Athens": {
        "artist": "Raphael",
        "year": "1511",
        "description": "《雅典學院》是拉斐爾的名作，描繪了古希臘哲學家的集會場景。"
    },
    "The Arnolfini Portrait": {
        "artist": "Jan van Eyck",
        "year": "1434",
        "description": "《阿諾芬尼夫婦像》是揚·范·艾克的名作，展示了北方文藝復興的精緻細節。"
    },
    "Las Meninas": {
        "artist": "Diego Velázquez",
        "year": "1656",
        "description": "《宮女》是維拉斯奎茲的名作，描繪了西班牙宮廷的生活場景。"
    },
    "The Girl with a Red Hat": {
        "artist": "Johannes Vermeer",
        "year": "1665",
        "description": "《戴紅帽的少女》是維梅爾的作品，展示了荷蘭黃金時代的日常生活。"
    },
    "The Hay Wain": {
        "artist": "John Constable",
        "year": "1821",
        "description": "《乾草車》是康斯特勃的名作，描繪了英國鄉村的風景。"
    },
    "The Swing": {
        "artist": "Jean-Honoré Fragonard",
        "year": "1767",
        "description": "《秋千》是弗拉戈納爾的作品，表現18世紀法國洛可可風格的浪漫主題。"
    },
    "The Death of Sardanapalus": {
        "artist": "Eugène Delacroix",
        "year": "1827",
        "description": "《薩達納帕魯斯之死》是德拉克羅瓦的作品，描繪了東方暴君的死亡場景。"
    },
    "Olympia": {
        "artist": "Édouard Manet",
        "year": "1863",
        "description": "《奧林匹亞》是馬奈的作品，挑戰了當時對女性裸體的傳統觀念。"
    }
}

# 上傳圖片
uploaded = files.upload()

# 處理圖片
for fname in uploaded.keys():
    image = Image.open(fname).convert("RGB")
    image_input = preprocess(image).unsqueeze(0).to(device)

    # 建立文字輸入
    painting_names = list(known_paintings.keys())
    text_inputs = clip.tokenize(painting_names).to(device)

    # 用 CLIP 做匹配
    with torch.no_grad():
        image_features = model.encode_image(image_input)
        text_features = model.encode_text(text_inputs)

        # 計算相似度
        similarity = (image_features @ text_features.T).squeeze(0)
        best_match_idx = similarity.argmax().item()
        matched_title = painting_names[best_match_idx]

    # 取得對應資訊
    info = known_paintings[matched_title]
    full_text = f"這幅畫是《{matched_title}》，由 {info['artist']} 在 {info['year']} 年創作。{info['description']}"
    print(full_text)

    # 語音播放
    tts = gTTS(full_text, lang='zh-tw')
    audio_path = f"/tmp/{uuid.uuid4()}.mp3"
    tts.save(audio_path)
    display(Audio(audio_path, autoplay=True))
```

```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```
